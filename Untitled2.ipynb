```python




def fan(row):
    if row['sex']=="female":
        return row["fare"]
    else:
        return -row["fare"]

df['fare']=df.apply(fan,axis=1)
g=sns.barplot(data=df,x='sex',y='fare',palette=['r','b'],hue='alive')
g.axhline(0,color='black',linestyle=":",linewidth=1.4)
g.set_xlabel('')
plt.show()
```
